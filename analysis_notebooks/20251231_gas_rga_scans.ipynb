```python
def load_rga_scan(file):
    out = np.loadtxt(file, delimiter=',', skiprows=27)

    amu = out.T[0]
    pressure_torr = out.T[1]
    pressure_mbar = pressure_torr * 1.33322
    return amu, pressure_mbar
```

```python
file_folder = r'/Users/yuhan/work/nanospheres/data/gas_rga_scans/scan_text_files'
xe_files = ['20251216_rga_scans_noxenon.txt',
            '20251216_rga_scans_xenon_cryopumped_fullrange.txt',
            '20251216_rga_scans_xenon_5e-8mbar_fullrange.txt',
            '20251216_rga_scans_xenon_8e-8mbar_fullrange.txt',
            '20251216_rga_scans_xenon_allrange.txt']

amus, pressures_xe = [], []
for file in xe_files:
    amu, p_mbar = load_rga_scan(f'{file_folder}/{file}')
    amus.append(amu)
    pressures_xe.append(p_mbar)
```

```python
labels= ['No injection', 'Cryopumped', '5e-8 mbar', '8e-8 mbar', '2e-7 mbar']

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for i in range(5):
    axes[0].plot(amus[i], pressures_xe[i], alpha=0.75, label=labels[i])
    axes[1].plot(amus[i], pressures_xe[i], alpha=0.75, label=labels[i])


axes[1].legend()
axes[0].set_xlim(0, 150)
axes[1].set_xlim(120, 140)

axes[0].set_ylim(1e-12, 1e-7)
axes[1].set_ylim(1e-12, 1e-9)

for ax in axes:
    ax.set_yscale('log')

    ax.set_xlabel('Atomic mass (amu)')
    ax.set_ylabel('Partial pressure (mbar)')

axes[1].text(128, 4e-11, r'$^{128}$Xe')
axes[1].text(129, 5e-10, r'$^{129}$Xe')
axes[1].text(130, 7e-11, r'$^{130}$Xe')
axes[1].text(131, 3.5e-10, r'$^{131}$Xe')
axes[1].text(132, 5e-10, r'$^{132}$Xe')
axes[1].text(134, 1.5e-10, r'$^{134}$Xe')
axes[1].text(136, 1e-10, r'$^{136}$Xe')

fig.tight_layout()
```
